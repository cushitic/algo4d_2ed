## Levenshtein Distance

```python
s1 = 'Saturday'
s2 = 'Sunday'
m = len(s1)
n = len(s2)
D = [list(range(n+1))]
for i in range(m):
        D.append([i+1] + [0] * n)
```

```python
for j in range(1, n+1):
    for i in range(1, m+1):
        if s1[i-1] == s2[j-1]: 
            D[i][j] = D[i-1][j-1]
        else:
            deletion = D[i-1][j] + 1
            insertion = D[i][j-1] + 1
            substitution = D[i-1][j-1] + 1
            D[i][j] = min(deletion, 
                          insertion,
                         substitution)
print (f'Levenshtein distance is {D[-1][-1]}')
```

```python
import pandas as pd

pd.DataFrame(D, columns=list(' '+s2), index=list(' '+s1))
```
