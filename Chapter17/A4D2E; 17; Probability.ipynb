# Representing statistical distributions

```python
from random import seed, gauss, uniform
import matplotlib.pyplot as plt
```

```python
seed(0)
normal_distribution = [gauss(mu=25, sigma=100) for r in range(10_000)]
weights = [1./10_000] * 10_000

plt.figure(figsize=(12, 6))
plt.hist(normal_distribution, bins=20, weights=weights,
         color='lightblue', edgecolor='black', linewidth=1.2)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
seed(0)
uniform_distribution = [uniform(a=0, b=100) for r in range(10_000)]
weights = [1./10_000] * 10_000

plt.figure(figsize=(12, 6))
plt.hist(uniform_distribution, bins=20, weights=weights,
         color='lightblue', edgecolor='black', linewidth=1.2)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
numbers = ['Ace','2','3','4','5','6','7','8','9','10',
                                'Jack','Queen','King']
seeds = ['Clubs','Spades','Diamonds','Hearts']
deck = [s+'_'+n for n in numbers for s in seeds]
```

```python
from random import choice

seed(0)
my_cards = deck.copy()
guessed = 0
for card in deck:
    if card == choice(my_cards):
        guessed += 1
print('Guessed %i card(s)' % guessed)
```

```python
seed(0)
my_cards = deck.copy()
guessed = 0
for card in deck:
    if card == choice(my_cards):
        guessed += 1
    else:
        my_cards.pop(my_cards.index(card))
print('Guessed %i card(s)' % guessed)
```

## Monte Carlo method

```python
seed(0)
samples = list()
for trial in range(10_000):
    my_cards = deck.copy()
    guessed = 0
    for card in deck:
        if card == choice(my_cards):
            guessed += 1
        else:
            my_cards.pop(my_cards.index(card))
    samples.append(guessed)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(samples, bins=8, color='lightblue', 
         edgecolor='black', linewidth=1.2)
plt.xlabel("Guesses")
plt.ylabel("Frequency")
plt.show()
mean = round(sum(samples) / len(samples), 2)
print(f'On average you can expect {mean} guesses each run')
```
